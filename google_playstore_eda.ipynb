# convert whole size column into bytes

```python
# for this, lets create a function

def convert_size(size):
    if isinstance(size, str):
        if 'k' in size:
            return float(size.replace('k',"")) * 1024
        elif 'M' in size:
            return float(size.replace('M',"")) * 1024 * 1024
        elif 'Varies with device' in size:
            return np.nan
    return size
```

#### apply this function

```python
df['Size'] = df['Size'].apply(convert_size)
```

```python
#first of all check in which installs category, rating has NaN values

df['Installs_category'].loc[df['Rating'].isnull()].value_counts()
```

#### lets fill the missing values

```python
# Calculate the mean rating for each Installs_category
mean_ratings = df.groupby('Installs_category')['Rating'].mean()

# Fill NaN values in the Rating column with the mean rating for each Installs_category
df['Rating'].fillna(df['Installs_category'].map(mean_ratings), inplace=True)
```
